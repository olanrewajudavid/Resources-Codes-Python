```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the website to scrape
url = "https://quotes.toscrape.com/"

# Send a GET request to the website
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find and extract specific elements from the webpage
quotes = soup.find_all("span", class_="text")
authors = soup.find_all("small", class_="author")

#create or open the csv file for scraped quotes
file = open("scraped_quotes.csv", "w")
writer = csv.writer(file)

#create headers
writer.writerow(["QUOTES", "AUTHORS"])

# Print the extracted data
for quote, author in zip(quotes, authors):
    print("Quote:", quote.get_text())
    print("Author:", author.get_text())

    #get each row and write it to the csv file
    writer.writerow([quote.get_text(), author.get_text()])



file.close()
```

```python
def scrape_quotes_from_page(url):
  #send a GET request to the website
  response = requests.get(url)

  #create a beautiful soup object to parse the html content out of
  soup = BeautifulSoup(response.content, "html.parser")

  #find the specific elements from the website that i want to scrape
  quotes = soup.find_all("span", class_="text")
  authors = soup.find_all("small", class_="author")

  for quote, author in zip(quotes, authors):
    print("Quote: ", quote.get_text())
    print("Author: ", author.get_text())
    print()
```

```python
def scrape_quotes_with_pagination(base_url,num_pages):
    for page in range(1,num_pages + 1):
      url = f"{base_url}?page={page}"
      print(f"Scraping page {page}...")
      scrape_quotes_from_page(url)
```

```python
base_url = "https://quotes.toscrape.com/"
num_pages = 10
scrape_quotes_with_pagination(base_url,num_pages)
```
